```python
import numpy as np
import matplotlib.pyplot as plt

baseline = np.loadtxt("results/baseline/mAP75.csv")
aug = np.loadtxt("results/da/mAP75.csv")


a = np.arange(1,51)
plt.figure(figsize=(12,5))
plt.plot(a, baseline,label="baseline")
plt.plot(a, aug,label="augmentation")

plt.legend(bbox_to_anchor=(1.0, 1), loc=0, borderaxespad=0.)
plt.xlabel('epochs')
plt.ylabel('mAP(IoU=0.75)')
plt.show()
#plt.savefig('test.png')
```

```python
loss = np.loadtxt("result/baseline/train.csv", skiprows=1)
loss1 = np.loadtxt("result/augumentation/train.csv", skiprows=1)
loss2 = np.loadtxt("result/focal loss/train.csv", skiprows=1)
loss3 = np.loadtxt("result/AugFL/train.csv", skiprows=1)

a = np.arange(1,151)
plt.figure(figsize=(12,5))
plt.plot(a, loss,label="baseline")
plt.plot(a, loss1,label="aug")
plt.plot(a, loss2,label="FL")
plt.plot(a, loss3,label="aug+FL")

plt.legend(bbox_to_anchor=(1.0, 1), loc=0, borderaxespad=0.)
plt.xlabel('epochs')
plt.ylabel('training loss')
plt.show()
#plt.savefig('test.png')
```
